```python
def plot_curve(x,y1,y2=None,legend = None, mode='loss',lr=False):
    y_lim=15.0
    val_min = y2.idxmin()
    print(val_min)
    print(y2.min())
    plt.figure()
    plt.plot(x,y1,color='red')
    if y2 is not None:
        plt.plot(x,y2,color='blue')
        if mode == 'loss':
            plt.vlines(x = val_min, ymin = 0, ymax = y_lim, 
           colors = 'green', 
           label = 'Minimum validation loss') 
            legend.append('Min val loss')
    plt.legend(legend)
    

    plt.xlabel('Epochs')
    plt.ylabel('%s' %('Loss' if mode=='loss' else 'MAE'))
    plt.title('%s' %('Loss' if mode=='loss' else 'MAE'))
    plt.grid()
    plt.ylim(0.0,y_lim)
    plt.show()
```
